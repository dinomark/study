```python
gunku=df.loc[:,['설치시군구명']]
gunku
```

```python
gunku=gunku.groupby('설치시군구명',as_index=False).size()
```

```python
gunku
```

```python
gunku.columns=['구','개수']
gunku
```

```python
gunku_num=np.array(gunku['개수'])
gunku_num
gunku_ku=np.array(gunku['구'])
gunku_ku
```

```python
plt.rc('font',family='Malgun Gothic')
plt.axis('equal')
plt.pie(gunku_num,labels=gunku_ku,autopct='%1.f%%',startangle=90)
plt.title('행정구별 와이파이 비율')
plt.show()
```
