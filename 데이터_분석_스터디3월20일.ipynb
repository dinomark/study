```python
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
print("iris의 key:{}".format(iris.keys()))#format=iris의 key형식 데이터 삽입
print(iris['DESCR'])#describe=descr
```

```python
print("target 이름:{}".format(iris['target_names']))
print("data의 타입:{}".format(type(iris['data'])))
print("data의 크기:{}".format(iris['data'].shape))
print("data의 처음 다섯 행:\n{}".format(iris['data'][:5]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(
    iris['data'],iris['target'],random_state=0)
print("X_train 크기:{}".format(X_train.shape))
print("X_test 크기:{}".format(X_test.shape))
```

```python
iris_dataframe=pd.DataFrame(X_train, columns=iris.feature_names)
print('#iris describe()=',iris_dataframe.describe())
```

```python
ax=pd.plotting.scatter_matrix(iris_dataframe,c=y_train, figsize=(15,15),marker='o',
                             hist_kwds={'bins':20},s=60,alpha=8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1) #1-NN
knn.fit(X_train, y_train)
```

```python
X_new=np.array([[5,2.9,1,0.2]])#머신러닝은 무조건 2차원 배열
print("X_new.shape:{}".format(X_new.shape))

prediction=knn.predict(X_new)
print("예측:{}".format(prediction))
print("예측한 타깃의 이름:{}".format(
    iris['target_names'][prediction]))
```
